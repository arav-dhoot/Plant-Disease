```python
import os
import csv
```

```python
lines = list()
with open('train.csv') as file:
    for line in file.readlines():
        lines.append(line.split(","))
lines = lines[1:]
write_list = list()
for line in lines:
    state = ""
    img_name = line[0]
    if line[1] == "1":
        state = "Healthy"
    elif line[2] == "1":
        state = "Multiple_Disease"
    elif line[3] == "1":
        state = "Rust"
    else:
        state = "Scab"
    write_list.append([f'{img_name}.jpg',state])
write_list
```

```python
with open('plant_pathology_train.csv', 'w') as csv_writer:
    writer = csv.writer(csv_writer)
    for images in write_list:
        writer.writerow(images)
```
