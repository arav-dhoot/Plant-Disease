```python
import os
import csv
```

```python
train_dir = "/Users/aravdhoot/CS_Research_Project/data/Plant Disease Recognition/Train/Train"
valid_dir = "/Users/aravdhoot/CS_Research_Project/data/Plant Disease Recognition/Validation/Validation"
test_dir = "/Users/aravdhoot/CS_Research_Project/data/Plant Disease Recognition/Test/Test"
```

```python
images_list = []
for file in os.listdir(train_dir):
    f = os.path.join(train_dir, file)
    for img in os.listdir(f):
        images_list.append([img, file])
images_list   
```

```python
with open("plants_disease__recognition_train_with_labels.csv", 'w') as csv_writer:
    writer = csv.writer(csv_writer)
    for images in images_list:
        writer.writerow(images)
```

```python
with open("plants_disease__recognition_valid_with_labels.csv", 'w') as csv_writer:
    writer = csv.writer(csv_writer)
    for images in images_list:
        writer.writerow(images)
```
